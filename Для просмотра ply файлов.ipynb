```python
import os
import open3d as o3d

def list_files_in_current_directory():
    print("üìÅ –°–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ç–µ–∫—É—â–µ–π –ø–∞–ø–∫–∏:\n")
    files = os.listdir(".")
    for f in files:
        print(f" - {f}")
    print("\n")

def view_ply_file(filename):
    if not os.path.isfile(filename):
        print(f"‚ùå –§–∞–π–ª '{filename}' –Ω–µ –Ω–∞–π–¥–µ–Ω.")
        return

    if not filename.lower().endswith(".ply"):
        print("‚ö†Ô∏è –≠—Ç–æ –Ω–µ .ply —Ñ–∞–π–ª.")
        return

    print(f"üîç –û—Ç–∫—Ä—ã–≤–∞—é {filename} ...")
    
    # –°–Ω–∞—á–∞–ª–∞ –ø—Ä–æ–±—É–µ–º –æ—Ç–∫—Ä—ã—Ç—å –∫–∞–∫ mesh
    mesh = o3d.io.read_triangle_mesh(filename)
    
    if mesh.has_triangles():
        # –≠—Ç–æ mesh —Å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞–º–∏
        print("‚úÖ –§–∞–π–ª —Å–æ–¥–µ—Ä–∂–∏—Ç mesh (—Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∏)")
        mesh.compute_vertex_normals()
        o3d.visualization.draw_geometries([mesh])
    elif mesh.has_vertices():
        # –≠—Ç–æ point cloud (—Ç–æ–ª—å–∫–æ —Ç–æ—á–∫–∏)
        print("‚úÖ –§–∞–π–ª —Å–æ–¥–µ—Ä–∂–∏—Ç point cloud (—Ç–æ–ª—å–∫–æ —Ç–æ—á–∫–∏)")
        pcd = o3d.io.read_point_cloud(filename)
        o3d.visualization.draw_geometries([pcd])
    else:
        print("‚ùå –û—à–∏–±–∫–∞: —Ñ–∞–π–ª –Ω–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–∞–Ω–Ω—ã—Ö.")

if __name__ == "__main__":
    list_files_in_current_directory()
    filename = input("–í–≤–µ–¥–∏—Ç–µ –∏–º—è .ply —Ñ–∞–π–ª–∞ –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞: ").strip()
    view_ply_file(filename)
```

```python
import os
import open3d as o3d
import numpy as np

def quick_check_ply(filename):
    """–ë—ã—Å—Ç—Ä–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –±–µ–∑ plyfile"""
    
    # –†–∞–∑–º–µ—Ä —Ñ–∞–π–ª–∞
    file_size = os.path.getsize(filename) / (1024 * 1024)
    print(f"\nüîç –ê–Ω–∞–ª–∏–∑: {filename}")
    print(f"üì¶ –†–∞–∑–º–µ—Ä: {file_size:.2f} MB\n")
    
    print("‚è≥ –ó–∞–≥—Ä—É–∑–∫–∞...")
    pcd = o3d.io.read_point_cloud(filename)
    print("‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω–æ!\n")
    
    # –ë–∞–∑–æ–≤–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
    num_points = len(pcd.points)
    print(f"üìä –¢–æ—á–µ–∫: {num_points:,}")
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–∑–º–µ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    print(f"\nüé® –†–∞–∑–º–µ—á–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
    print(f"   –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã (x,y,z): ‚úÖ –î–∞")
    print(f"   –¶–≤–µ—Ç–∞ (RGB): {'‚úÖ –î–∞' if pcd.has_colors() else '‚ùå –ù–µ—Ç'}")
    print(f"   –ù–æ—Ä–º–∞–ª–∏ (nx,ny,nz): {'‚úÖ –î–∞' if pcd.has_normals() else '‚ùå –ù–µ—Ç'}")
    
    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö
    if pcd.has_colors():
        colors = np.asarray(pcd.colors)
        print(f"\n   üìà –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ü–≤–µ—Ç–æ–≤:")
        print(f"      Min RGB: {colors.min(axis=0)}")
        print(f"      Max RGB: {colors.max(axis=0)}")
        print(f"      Mean RGB: {colors.mean(axis=0)}")
    
    if pcd.has_normals():
        normals = np.asarray(pcd.normals)
        print(f"\n   üìê –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –Ω–æ—Ä–º–∞–ª–µ–π:")
        print(f"      –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ: {len(normals)}")
    
    print(f"\n{'='*50}")
    print("‚úÖ –ü—Ä–æ–≤–µ—Ä–∫–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞")
    print(f"{'='*50}\n")
    
    return pcd

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
if __name__ == "__main__":
    filename = "L001.ply"
    
    if not os.path.isfile(filename):
        print(f"‚ùå –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω!")
    else:
        pcd = quick_check_ply(filename)
        
        # –û–ø—Ü–∏—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
        show = input("–û—Ç–∫—Ä—ã—Ç—å –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é? (y/n): ").strip().lower()
        if show == 'y':
            if not pcd.has_colors():
                pcd.paint_uniform_color([0.7, 0.7, 0.7])
            o3d.visualization.draw_geometries([pcd])
```
